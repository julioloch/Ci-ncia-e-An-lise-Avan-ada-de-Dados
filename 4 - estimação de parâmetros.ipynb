```python
import scipy.stats as st
nivel_de_confianca = 0.95
nivel_de_significancia = 1 - nivel_de_confianca
Z = abs(st.norm.ppf(nivel_de_significancia/2))
print(f'Z = {Z}')
```

```python
import numpy as np
nums = np.array([3, 3, 6, 5, 2, 2, 3, 10, 20, 20, 4, 3, 1, 2])
nums.sort()


mediana = np.median(nums)
media = np.mean(nums)

print(f"mediana: {mediana}, media {media}")
```
